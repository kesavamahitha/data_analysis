```python
import pandas as pd
```

```python


file_path = r"C:\Users\Mahitha\Desktop\Internship\data_superstore.csv"
df = pd.read_csv(file_path, encoding='windows-1252')
```

```python
df.shape
```

```python
df.columns
```

```python
columns_to_drop = ['Ship Date','Ship Mode', 'Customer Name', 'Segment', 'Country', 'City', 'Postal Code', 'Region', 'Product ID', 'Discount']
df.drop(columns=columns_to_drop, inplace=True)
df.columns
```

```python
import matplotlib.pyplot as plt
```

```python
grouped_df1 = df.groupby('Category')['Sales'].sum()
plt.figure(figsize=(8, 8))
plt.pie(grouped_df1, labels=grouped_df1.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution by Category')
plt.show()



```

```python
grouped_df2 = df.groupby('Category')['Profit'].sum()
plt.figure(figsize=(8, 8))
plt.pie(grouped_df2, labels=grouped_df2.index, autopct='%1.1f%%', startangle=90)
plt.title('Profit Distribution by Category')
plt.show()
```

```python
grouped_segment = df.groupby('State').sum()[['Sales', 'Profit']]

grouped_segment.plot(kind='bar', figsize=(10, 6), title='Total Sales and Profit by State')
plt.xlabel('State')
plt.ylabel('Value')
plt.show()
```

```python
top_products_quantity = df.groupby('Product Name')['Quantity'].sum().nlargest(10)
top_products_quantity.plot(kind='bar', figsize=(12, 6), title='Top 10 Products by Quantity Sold')
plt.xlabel('Product Name')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()

```
